```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
import pandas as pd
```

```python
feedback_df = pd.read_csv('insurance_review_cleaned.csv')
```

```python
feedback_df
```

```python
list(feedback_df)
```

```python
feedback_df = feedback_df[feedback_df['Insurance_experience'] != ' '].dropna(subset=['Insurance_experience'])
feedback_df
```

```python
data = feedback_df['Insurance_experience']
data.dtype
```

```python
data = list(data)
print(data)
```

```python
sentiment_pipeline(data)
```

```python
result_lst = sentiment_pipeline(data)
```

```python
result_df = pd.DataFrame(result_lst) 
result_df
```

```python
df_t = []
for ind in result_df.index:
    if result_df['label'][ind] == 'POSITIVE':
        df_t.append(1)
    elif result_df['label'][ind] == 'NEGATIVE':
        df_t.append(0)
```

```python
df_t
```

```python
df_t = pd.DataFrame(df_t, columns=['score_round'])
```

```python
df_t
```

```python
final_df = pd.merge(feedback_df['Insurance_experience'], result_df, left_index=True, right_index=True, how='outer')
```

```python
final_df
```

```python
final_df = pd.merge(final_df, df_t, left_index=True, right_index=True, how='outer')
```

```python
final_df
```

```python
final_df.to_csv('Final_Result.csv',index=False)
```
