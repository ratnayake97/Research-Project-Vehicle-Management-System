```python
import pandas as pd
import pickle
```

```python
# Import insurance data file reviewed by the customers
insurance_df = pd.read_csv('Customer_Feedback.csv')
insurance_df
```

```python
# Remove the string time from Timestamp, using .str.slice(start,stop,step) method
insurance_df['Timestamp'] = insurance_df['Timestamp'].str.slice(start=None,stop=11,step=1)
insurance_df
```

```python
#Rename the Columns
insurance_df.rename(columns={"Timestamp": "Date","What is the Motor Insurance are you using?": "Motor_Insurance","Vehicle Type": "Vehicle_Type","Vehicle Make": "Vehicle_Make","Vehicle Model": "Vehicle_Model","How long have you been using Insurance Plan?": "Years_of_experience","Provide honest opinion regarding the motor insurance experience you are using?": "Insurance_experience","Rate your Insurance": "Rate"}, inplace=True)
insurance_df
```

```python
# Merge 2 two data in to one column
insurance_df['Vehicle'] =insurance_df['Vehicle_Make'].astype(str) + ' ' + insurance_df['Vehicle_Model']
insurance_df
```

```python
# Covert the Motor Insurance  column into uppercase
insurance_df['Motor_Insurance'] = insurance_df['Motor_Insurance'].str.upper()

# Covert the Vehicle Type column into uppercase
insurance_df['Vehicle_Type'] = insurance_df['Vehicle_Type'].str.upper()

# Covert the Vehicle Make column into uppercase
insurance_df['Vehicle_Make'] = insurance_df['Vehicle_Make'].str.upper()

# Covert the Vehicle Model column into uppercase
insurance_df['Vehicle_Model'] = insurance_df['Vehicle_Model'].str.upper()

# Covert the Insurance experience column into uppercase
insurance_df['Insurance_experience'] = insurance_df['Insurance_experience'].str.upper()

# Covert the Vehicle column into uppercase
insurance_df['Vehicle'] = insurance_df['Vehicle'].str.upper()
insurance_df
```

```python
# Write the Edited insurance reviews file to .csv
insurance_df.to_csv('insurance_review_cleaned.csv',index=False)
```

```python
insurance_df
```

```python
# Find the no of vehicle for each Insurance
insu_veh = insurance_df.groupby(['Motor_Insurance','Vehicle']).agg({'Vehicle':'nunique'})
```

```python
insu_veh
```

```python
insu_veh = insu_veh.rename(columns={'Vehicle':'Qty'})
```

```python
insu_veh
```

```python
# Reset the index by converting the Insurance into a column
insu_veh.reset_index(inplace=True)
```

```python
insu_veh
```

```python
insu_veh.to_csv('insu_veh.csv')
```

```python
# Create a pivot table with all vehical type on columns and Insurance on rows, and Qty as values
insu_veh_pivot_df = insu_veh.pivot(index='Motor_Insurance',columns='Vehicle',values='Qty').fillna(0)
```

```python
insu_veh_pivot_df
```

```python
# Find the correlation between every two Vehicles and build a correlation matrix using corr() method
# Used Spearman method in identifying the correlation. Pearson was not providing better results and Kendall is taking a long time for execution.
insu_veh_correlation_df = insu_veh_pivot_df.corr(method='spearman',min_periods=5)
```

```python
insu_veh_correlation_df
```

```python
insu_veh_correlation_df.to_csv('Insurance_Vehical_Correlation_Matrix.csv')
```

```python
pickle.dump(insu_veh_correlation_df, open('Insurance_Vehical_correlation_model.pkl','wb'))
```

```python
# Sum of ratings for each insurance
insu_rate = insurance_df.groupby('Motor_Insurance').agg({'Rate':'sum'})
```

```python
insu_rate
```

```python
insu_rate.reset_index(inplace=True)
```

```python
insu_rate
```

```python
# Rank the Insurance according to the ratings
insu_rate['Tot_rate_insu'] = insu_rate['Rate'].rank(method='min', ascending=False).astype(int)
```

```python
insu_rate
```

```python
# List top ten insurance
insu_rate.sort_values('Tot_rate_insu').head(10)
```

```python
# Find unique total insurance and ratings for each one
# get total ratings
# get insurance count
insu_rate = insurance_df.groupby('Motor_Insurance').agg({'Rate':'sum', 'Motor_Insurance' : 'count'})
```

```python
insu_rate
```

```python
insu_rate = insu_rate.rename(columns={'Motor_Insurance':'num_count'})
```

```python
insu_rate.reset_index(inplace=True)
```

```python
insu_rate
```

```python
# average ratings for insurances
# average_ratings = total_ratins / frequency
tot_rate = insu_rate['Rate']
tot_count_insu = insu_rate['num_count']
```

```python
insu_rate['weight'] = tot_rate / tot_count_insu
```

```python
insu_rate
```

```python
# rank insurance
insu_rate['Top_rate_insu'] = insu_rate['weight'].rank(method='min', ascending=False).astype(int)
```

```python
insu_rate
```

```python
# The insurance with the highest rating average
insu_rate.sort_values('Top_rate_insu').head(10)
```
