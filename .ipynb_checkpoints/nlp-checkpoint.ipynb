```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
data = ["I love you", "I hate you", "Worst experience", "Still i didn't used, i think it's good so, i give 5 stars"]
sentiment_pipeline(data)
```

```python
import pandas as pd
```

```python
feedback_df = pd.read_csv('Customer_Feedback.csv')
```

```python
feedback_df
```

```python
list(feedback_df)
```

```python
feedback_df.rename({"Provide honest opinion regarding the motor insurance experience you are using?": "Feedback"}, axis=1, inplace=True)
list(feedback_df)
```

```python
feedback_df = feedback_df[feedback_df['Feedback'] != ' '].dropna(subset=['Feedback'])
feedback_df
```

```python
data = feedback_df['Feedback']
data.dtype
```

```python
data = list(data)
print(data)
```

```python
sentiment_pipeline(data)
```

```python
result_lst = sentiment_pipeline(data)
```

```python
result_df = pd.DataFrame(result_lst) 
result_df
```

```python
result_df = pd.merge(feedback_df['Feedback'], result_df, left_index=True, right_index=True, how='outer')
```

```python
result_df
```

```python
result_df.to_csv('Final_Result.csv',index=False)
```
