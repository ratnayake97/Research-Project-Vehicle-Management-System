# Top Rating Insurance

```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
# Import Feature Engineered data file
df = pd.read_csv('insurance_review_cleaned.csv')
df
```

```python
# Sum of ratings for each insurance
top_rating_insurace = df.groupby('Motor_Insurance').agg({'Rate':'sum'})
top_rating_insurace.head()
```

```python
top_rating_insurace.reset_index(inplace=True)
top_rating_insurace.head()
```

```python
# Rank the Insurance according to the ratings
top_rating_insurace['Total_rating'] = top_rating_insurace['Rate'].rank(method='min', ascending=False).astype(int)
top_rating_insurace.head()
```

```python
# List top ten insurance
top_rating_insurace.sort_values('Total_rating').head(10)
```

```python
# Find unique total insurance and ratings for each one
# get total ratings
# get insurance count
top_average_rating_insurace = df.groupby('Motor_Insurance').agg({'Rate':'sum', 'Motor_Insurance' : 'count'})
top_average_rating_insurace
```

```python
top_average_rating_insurace = top_average_rating_insurace.rename(columns={'Motor_Insurance':'num_count'})
top_average_rating_insurace.reset_index(inplace=True)
top_average_rating_insurace.head()
```

```python
# average ratings for insurances
# average_ratings = total_ratins / frequency
tot_rate = top_average_rating_insurace['Rate']
tot_count = top_average_rating_insurace['num_count']
top_average_rating_insurace['weight'] = tot_rate / tot_count
```

```python
top_average_rating_insurace
```

```python
# rank insurance
top_average_rating_insurace['average_rate_insurance'] = top_average_rating_insurace['weight'].rank(method='min', ascending=False).astype(int)
top_average_rating_insurace
```

```python
# The insurance with the highest rating average
top_average_rating_insurace.sort_values('average_rate_insurance').head(10)
```

```python
#Top 10 average rating insurance
top_average_rating_insurace10 = top_average_rating_insurace.sort_values("average_rate_insurance").head(10)
top_average_rating_insurace10
```

```python
#Write the Average Rankings into a .csv file
top_average_rating_insurace10.to_csv('CSV/top_average_rating_insurance.csv',index=False)
# Create a Pickle (.pkl) file with the average Ranking dataframe
pickle.dump(top_average_rating_insurace10, open('Models/top_average_rating_insurance.pkl','wb'))
```
