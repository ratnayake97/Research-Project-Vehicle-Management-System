```python
import pandas as pd
import pickle
```

```python
# Import insurance data file reviewed by the customers
insurance_df = pd.read_csv('insurance_review_cleaned.csv')
insurance_df
```

```python
# Find the no of vehicle for each Insurance
insu_veh = insurance_df.groupby(['Motor_Insurance','Vehicle']).agg({'Vehicle':'nunique'})
```

```python
insu_veh
```

```python
insu_veh = insu_veh.rename(columns={'Vehicle':'Qty'})
```

```python
insu_veh
```

```python
# Reset the index by converting the Insurance into a column
insu_veh.reset_index(inplace=True)
```

```python
insu_veh
```

```python
insu_veh.to_csv('insu_veh.csv')
```

```python
# Create a pivot table with all vehical type on columns and Insurance on rows, and Qty as values
insu_veh_pivot_df = insu_veh.pivot(index='Motor_Insurance',columns='Vehicle',values='Qty').fillna(0)
```

```python
insu_veh_pivot_df
```

```python
# Find the correlation between every two Vehicles and build a correlation matrix using corr() method
# Used Spearman method in identifying the correlation. Pearson was not providing better results and Kendall is taking a long time for execution.
insu_veh_correlation_df = insu_veh_pivot_df.corr(method='spearman',min_periods=5)
```

```python
insu_veh_correlation_df
```

```python
insu_veh_correlation_df.to_csv('Insurance_Vehical_Correlation_Matrix.csv')
```

```python
pickle.dump(insu_veh_correlation_df, open('Insurance_Vehical_correlation_model.pkl','wb'))
```

```python
# Sum of ratings for each insurance
insu_rate = insurance_df.groupby('Motor_Insurance').agg({'Rate':'sum'})
```

```python
insu_rate
```

```python
insu_rate.reset_index(inplace=True)
```

```python
insu_rate
```

```python
# Rank the Insurance according to the ratings
insu_rate['Tot_rate_insu'] = insu_rate['Rate'].rank(method='min', ascending=False).astype(int)
```

```python
insu_rate
```

```python
# List top ten insurance
insu_rate.sort_values('Tot_rate_insu').head(10)
```

```python
# Find unique total insurance and ratings for each one
# get total ratings
# get insurance count
insu_rate = insurance_df.groupby('Motor_Insurance').agg({'Rate':'sum', 'Motor_Insurance' : 'count'})
```

```python
insu_rate
```

```python
insu_rate = insu_rate.rename(columns={'Motor_Insurance':'num_count'})
```

```python
insu_rate.reset_index(inplace=True)
```

```python
insu_rate
```

```python
# average ratings for insurances
# average_ratings = total_ratins / frequency
tot_rate = insu_rate['Rate']
tot_count_insu = insu_rate['num_count']
```

```python
insu_rate['weight'] = tot_rate / tot_count_insu
```

```python
insu_rate
```

```python
# rank insurance
insu_rate['Top_rate_insu'] = insu_rate['weight'].rank(method='min', ascending=False).astype(int)
```

```python
insu_rate
```

```python
# The insurance with the highest rating average
insu_rate.sort_values('Top_rate_insu').head(10)
```
