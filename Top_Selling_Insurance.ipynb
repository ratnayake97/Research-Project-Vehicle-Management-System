# Top Selling Insurance

```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
# Import Feature Engineered data file
df = pd.read_csv('insurance_review_cleaned.csv')
df
```

```python
# Find the no of vehicle for each Insurance
top_insurance = df.groupby(['Motor_Insurance']).agg({'Vehicle':'count'})
top_insurance
```

```python
# Reset the index by converting the Motor_Insurance into a column
top_insurance.reset_index(inplace=True)
top_insurance
```

```python
#Rename the Column
top_insurance.rename(columns={"Vehicle": "Vehicle_Count"}, inplace=True)
top_insurance
```

```python
# Rank the Motor_Insurance by most number of sold
top_insurance['Top_Sell_Insurance'] = top_insurance['Vehicle_Count'].rank(method='min',ascending=False).astype(int)
top_insurance
```

```python
# top selling insurance
top_insurance10 = top_insurance.sort_values("Top_Sell_Insurance").head(10)
top_insurance10
```

```python
#Write the Rankings into a .csv file
top_insurance10.to_csv('top_insurance.csv',index=False)
# Create a Pickle (.pkl) file with the Ranking dataframe
pickle.dump(top_insurance10, open('Models/top_insurance.pkl','wb'))
```
