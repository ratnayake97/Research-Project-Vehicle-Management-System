```python
# Import insurance data file reviewed by the customers
insurance_df = pd.read_csv('Insurance_Feedback.csv')
insurance_df
```

```python
# Remove the string time from Timestamp, using .str.slice(start,stop,step) method
insurance_df['Timestamp'] = insurance_df['Timestamp'].str.slice(start=None,stop=11,step=1)
insurance_df
```

```python
#Rename the Columns
insurance_df.rename(columns={"Timestamp": "Date","What is the Motor Insurance are you using?": "Motor_Insurance","Vehicle Type": "Vehicle_Type","Vehicle Make": "Vehicle_Make","Vehicle Model": "Vehicle_Model","How long have you been using Insurance Plan?": "Years_of_experience","Provide honest opinion regarding the motor insurance experience you are using?": "Insurance_experience","Rate your Insurance": "Rate"}, inplace=True)
insurance_df
```

```python
# Merge 2 two data in to one column
insurance_df['Vehicle'] =insurance_df['Vehicle_Make'].astype(str) + ' ' + insurance_df['Vehicle_Model']
insurance_df
```

```python
# Covert the Motor Insurance  column into uppercase
insurance_df['Motor_Insurance'] = insurance_df['Motor_Insurance'].str.upper()

# Covert the Vehicle Type column into uppercase
insurance_df['Vehicle_Type'] = insurance_df['Vehicle_Type'].str.upper()

# Covert the Vehicle Make column into uppercase
insurance_df['Vehicle_Make'] = insurance_df['Vehicle_Make'].str.upper()

# Covert the Vehicle Model column into uppercase
insurance_df['Vehicle_Model'] = insurance_df['Vehicle_Model'].str.upper()

# Covert the Insurance experience column into uppercase
insurance_df['Insurance_experience'] = insurance_df['Insurance_experience'].str.upper()

# Covert the Vehicle column into uppercase
insurance_df['Vehicle'] = insurance_df['Vehicle'].str.upper()
insurance_df
```

```python
# Write the Edited insurance reviews file to .csv
insurance_df.to_csv('insurance_review_cleaned.csv',index=False)
```
